## Imports

```python
import tensorflow as tf
import gpt_2_simple as gpt2
import os
import re
import pandas as pd
```

## Functions

```python
def read_txt_file(data_file):
    record = []
    with open(data_file, 'r') as f:
        for line in f:
            
            record.append(line)
    
    return record
```

```python
def generate(data_row, run_name, temp, nsamples, batch_size):
    
    tf.reset_default_graph()
    session = gpt2.start_tf_sess()
    
    test_lst = data_row.split('<|endoftext|>\n')
    test_str = test_lst[0].split('[QUESTION]:')[0]
    
    pre = test_str + " [QUESTION]:"
    
    gpt2.load_gpt2(session, run_name = run_name)
    
    generated = gpt2.generate(session,
                          temperature = temp,
                          prefix= pre,
                          nsamples = nsamples,
                          batch_size = batch_size,
                          run_name = run_name,
                          return_as_list=True
                        )
    return generated 
```

## Generation

```python
data_file = 'Data/quac_test.txt'
```

```python
records = read_txt_file(data_file)
```

```python
eval_df = pd.DataFrame(records, columns = ['Context'])
```

```python
eval_df
```

```python
# call generate function 

run_name = 'quac_ht_1' 
temp = 0.7 
nsamples = 5 
batch_size = 5

question_lst = [] 

for idx, row in eval_df.iterrows():
    print(idx)
    gen_lst = generate(row['Context'], run_name, temp, nsamples, batch_size)
    question_lst.append(gen_lst)
```

```python
eval_df['Generated'] = question_lst
```

```python
eval_df
```

```python
question_col = []
for idx, item in eval_df.iterrows():
    questions = []
    item_lst = item['Generated'][0].split('<|startoftext|>')
    
    for string in item_lst[1:]:
        
        try:
            ques = re.search(r'\[QUESTION\]\:(.+?)<\|endoftext\|>', string).group(1)
            questions.append(ques)
        except:
            questions.append('None')
            print("No String")

    question_col.append(questions)
```

```python
eval_df['Questions'] = question_col
```

```python
eval_df
```

```python
context_clean = []
for idx, item in eval_df.iterrows():
    string = item['Context']
    try:
        con = re.search(r'\[CONTEXT\]\: (.+?)\[QUESTION\]\:', string).group(1)
        context_clean.append(con)
    except:
        print("No String")
        print(string)
```

```python
final_questions = []
final_context = []

for idx, row in eval_df.iterrows():
    for item in row['Questions']:
        final_context.append(context_clean[idx])
        final_questions.append(item)
    
```

```python
final_eval = pd.DataFrame({'Context' : final_context, 'Question' : final_questions})
```
